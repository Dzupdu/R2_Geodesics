```python
from sympy.interactive import printing
printing.init_printing(use_latex=True)
from sympy.diffgeom.rn import R2
from sympy.diffgeom import TensorProduct
import matplotlib.pyplot as plt
TP = TensorProduct

# self defined functions for actual calculation
from R2_Geodesics import metric_to_geodesic_eq,solve_geodesic_eq, plot_geodesics
```

```python
# define metric, must be symmetric
g = R2.x*TP(R2.dx, R2.dx) + TP(R2.dy, R2.dy)
display(g)
```

```python
# symbolically find the geodesic equations
eqs = metric_to_geodesic_eq(g)
display(eqs)
```

```python
# solve geodesic equations numerically and plot 
# one initial point with different angles of initial velocity
plot_geodesics(eqs, [0.1, 0], tmax = 10)
```

```python
# same metric, mirrored starting point

plot_geodesics(eqs, [-0.1, 0.1], tmax = 10)
```

```python
# another, more complex metric
g2 = (-1/(1-R2.x*R2.y)**2)*TP(R2.dx, R2.dy) + (-1/(1-R2.x*R2.y)**2)*TP(R2.dy, R2.dx)
display(g2)
```

```python
eqs2 = metric_to_geodesic_eq(g2)
display(eqs2)
```

```python
# solve geodesic equations numerically and plot 
# one initial point with different angles of initial velocity
plot_geodesics(eqs2, [0.01, 0.01], xlim = 10, ylim =10, tmax = 10)
```

```python
#another starting point
plot_geodesics(eqs2, [-1, 1], xlim = 10, ylim =10, tmax = 10)
```

```python
#another starting point
plot_geodesics(eqs2, [-1, 2], xlim = 10, ylim =10, tmax = 10)
```
